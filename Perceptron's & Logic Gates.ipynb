**Perceptrons & Neural Networks to Artificially find the Boundary of an AND Logic Gate**

Building a simple artificial neural network that works on one perceptron to teach it the input and output of an AND & OR gate. This works because this can be considered to be linearly separable so a decision boundary is created by the perceptron algorithm. The visual colour plot shows the distance of different inputs to this boundary (Dark Purple line)

It is possible to use perceptrons to model the fundamental building blocks of computers â€” logic gates. For example, the table below shows the results of an AND gate. Given two inputs, an AND gate will output a 1 only if both inputs are a 1:

Perceptrons.JPG

This Perceptron model will aim to learn the AND & OR Gate and produce a 100% accuracy on the training data. Will also use it to create a heat map to show exactly where the Perceptron has drawn a line to linearly seperate the two different possible outcomes of the AND Gate: 1 or 0. (To test for the OR gate, just comment out code under AND GATE COMMENT & UNCOMMENT OR GATE CODE.


```python

from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product

#AND GATE
data = [[0,0], [1,1], [0,1], [1,0]]
labels = [0, 1, 0,0]

#OR GATE
#data = [ [0,0], [0,1], [1,0], [1,1] ]
#labels = [0,1,1,1]

validation = [[0, 0], [1, 1], [0.5, 0.5]]
classifier = Perceptron(max_iter = 40)
plt.scatter([point[0] for point in data], 
            [point[1] for point in data], 
            c=labels)
classifier.fit(data, labels)
print(classifier.score(data, labels))

x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)
point_grid = list(product(x_values,y_values))
distances = classifier.decision_function(point_grid)
abs_distances = [abs(x) for x in distances]
distances_matrix = np.reshape(abs_distances, (100,100))
heatmap = plt.pcolormesh(x_values, y_values, distances_matrix)
plt.colorbar(heatmap)

plt.show()

```
